# ***Image to Pencil Sketch using python***

**Step 1:** Include all the primary and necessary header files.

```python
import cv2
import matplotlib.pyplot as plt
```

**Step 2:** Load the Original image in which changes has to be made.

```python
img=cv2.imread("image.jpg")
img[:1]
```

For grayscale images, A two dimensional array is created and if the image contains color, a three dimensional array is created.

**Step 3:** OpenCV by default reads images in BGR format, to convert that into RGB we use cvtColor() method.

```python
rgb_img=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
rgb_img[:1]
```

Notice, The last and the first column values are exchanged to form a RGB image.

```python
fig ,(ax0, ax1) = plt.subplots(nrows=1,
                               ncols=2,
                               figsize= (10,6))

ax0.axis('off')
ax1.axis('off')

ax0.imshow(img)
ax0.set(title="BGR Image");
ax0.title.set_color("blue")

ax1.imshow(rgb_img)
ax1.set(title="RGB Image")
ax1.title.set_color("blue")


# Add a title to figure

fig.suptitle("Conversion from BGR to RGB image",
            fontsize =16,
            fontweight="bold",
            color="black");
```

**Step 4:** cconversion of BGR image into Grayscale Image.

```python
gray_img=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

#visualizing the image

plt.imshow(gray_img)
plt.axis('off')
plt.title("GrayScale Image").set_color("blue")
```

**Step 5:** Inverting GrayScaled Image.

What are the ways in which we can invert an image.

a)Subtracting from 255.
   
   example= 255-image.



b)using bitwise not operator.

  example= cv2.bitwise_not(image)

```python
invert_img= cv2.bitwise_not(gray_img)
#using bitwise not

plt.imshow(invert_img)
plt.axis('off')
plt.title("Inverted GrayScale Image").set_color("blue")
```

**Step 6:** Bluring the inverted Image.

```python
blur_img = cv2.GaussianBlur(invert_img, (21,21), 0) 

plt.imshow(blur_img)
plt.axis('off')
plt.title("Blurred Image").set_color("blue")
```

**Step 7:** Inverting the blurred image.

```python
invert_blur_img = cv2.bitwise_not(blur_img)

plt.imshow(invert_blur_img)
plt.axis('off')
plt.title("Inverted Blurred Image").set_color("blue")
```

**Step 7:** Generating Pencil Sketch

```python
sketch_img = cv2.divide(gray_img, invert_blur_img, scale=256.0)

plt.imshow(sketch_img)
plt.axis('off')
plt.title("Pencil Sketched Image").set_color("blue")
```

```python
cv2.imwrite('image.jpg', sketch_img)


# Loaded image will be in BGR format so converting it into RGB and displaying.

output_bgr_img = cv2.imread('image.jpg')
output_img = cv2.cvtColor(output_bgr_img, cv2.COLOR_BGR2RGB)

fig ,(ax0, ax1) = plt.subplots(nrows=1,ncols=2, figsize= (10,6))
ax0.axis('off')
ax1.axis('off')

ax0.imshow(rgb_img)
ax0.set(title="Original Image");
ax0.title.set_color("blue")

ax1.imshow(output_img)
ax1.set(title="Sketched Image")
ax1.title.set_color("blue")


# Add a title to figure

fig.suptitle("Original Image vs Sketched Image",
            fontsize =16,
            fontweight="bold",
            color="black");
```
